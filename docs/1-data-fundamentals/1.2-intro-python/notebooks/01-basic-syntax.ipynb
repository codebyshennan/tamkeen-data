## üî¢ Section 2: Data Types

Python has different types of data. Let's explore them!

````python
# Numbers
integer_num = 42
float_num = 3.14

# Strings
text = "Hello, Python!"

# Boolean
is_true = True

# Check the types
print(f"Integer: {integer_num} is type {type(integer_num)}")
print(f"Float: {float_num} is type {type(float_num)}")
print(f"String: {text} is type {type(text)}")
print(f"Boolean: {is_true} is type {type(is_true)}")
````

### üß™ Experiment: Type Conversion

````python
# Converting between types
num_as_string = "123"
string_to_num = int(num_as_string)

print(f"Original: {num_as_string} (type: {type(num_as_string)})")
print(f"Converted: {string_to_num} (type: {type(string_to_num)})")

# Try some math
result = string_to_num + 10
print(f"After adding 10: {result}")
````

## üßÆ Section 3: Mathematical Operations

````python
# Basic math operations
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus (remainder): {a} % {b} = {a % b}")
print(f"Exponent: {a} ** {b} = {a ** b}")
````

## üìù Section 4: String Operations

````python
# String manipulation
text = "   Python is Awesome!   "

print(f"Original: '{text}'")
print(f"Lowercase: '{text.lower()}'")
print(f"Uppercase: '{text.upper()}'")
print(f"Stripped: '{text.strip()}'")
print(f"Replaced: '{text.replace('Awesome', 'Amazing')}'")
print(f"Length: {len(text)} characters")
````
